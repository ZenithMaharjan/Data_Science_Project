```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler
```

```python
num_vars = ['Engine Size(L)', 'Cylinders', 'Fuel Consumption City (L/100 km)', 'Fuel Consumption Hwy (L/100 km)' , 'Fuel Consumption Comb (L/100 km)' , 'Fuel Consumption Comb (mpg)' , 'CO2 Emissions(g/km)']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
X_train = df_train.loc[ : , df_train.columns != 'CO2 Emissions(g/km)']
X_train
```

```python
Y_train = df_train['CO2 Emissions(g/km)']
Y_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
rfe = RFE(ln, n_features_to_select=10)
#running RFE
rfe = rfe.fit(X_train, Y_train)
```

```python
X_train.columns
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
X_train_rfe
```

```python
# Adding a constant variable
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
X_train_rfe
```
